```python
import copy

N=8

# Options # This line appears to be a comment or placeholder and doesn't execute code

def printSolution (board):
    for row in board:
        for i in range(N):
            # This part of the printSolution function seems to have an error in indexing row[1].
            # It should likely be checking the value at index i within the row.
            print("0" if row[i] else ".", end="")
        print()
    print()


def isSafe (board, row, col):
    for i in range(row):
        if board[i][col]:
            return False

    # Correcting indentation and potential syntax error in the following loops
    for i, j in zip(range(row - 1, -1, -1), range(col - 1, -1, -1)):
        if board[i][j]:
            return False

    for i, j in zip(range(row - 1, -1, -1), range(col + 1, N)):
        if board[i][j]:
            return False

    return True

def solve (board, row, solutions):
    # Correcting syntax error for equality check
    if row == N:
        solutions.append(copy.deepcopy(board))
        #printSolution(board) # Commenting out printSolution here as it's called at the end
        return

    for col in range(N):
        if isSafe (board, row, col):
            board[row][col] = 1
            solve (board, row + 1, solutions)
            board[row][col] = 0 # Backtrack

def eightQueens():
    board = [[0 for _ in range (N)] for _ in range (N) ] # Using _ for unused variable
    solutions = [] # Correcting syntax error for list creation
    solve (board, 0, solutions)
    # Correcting syntax error in f-string
    print(f"Total solutions found: {len(solutions)}")
    # Print all solutions after the recursive calls are complete
    for sol in solutions:
        printSolution(sol)


eightQueens()
```
